**Contexto**

Você trabalha numa escola e percebeu que, no horário do intervalo, a fila para comprar lanches na cantina é muito longa e desorganizada. Para resolver isso, sua ideia é orientar aos alunos cada uma terá a sua ordem pré-estabelecida para comprar seus lanches.

Você receberá os dados na estrutura de dicionário, sendo:
1. As chaves dos dicionários representam a matrícula dos alunos;
2. As notas foram armazenadas numa lista que é acessada através da matrícula dos alunos, conforme o exemplo abaixo:

# Critérios de avaliação (resultado esperado)

O grupo deverá enviar um código em python que atenda aos seguintes critérios:

1. Tratamento de exceções: o grupo deverá ordenar as matrículas, conforme a nota total (ou média). Entretanto, as listas de notas possuem tipos diferentes de variáveis. O grupo deverá utilizar tratamento de exceção para realizar o cálculo da nota total ou da métrica de cada aluno;

2. Ordenação: utilize algum dos critérios de ordenação vistos em aula, adaptando para que apresente as notas em ordem decrescente (maior nota até a menor nota);

* Usaremos a **exceção** para calcular as notas

'001' -> 312,

...

Portando, o resultado seria ['007', '004', '006', '001', '000', '005', '003', '009', '008', '002']
